```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# ─── Load Data ───────────────────────────────────────────────────────────────
dat = np.load('/home/maria/LuckyMouse4/data/logreg_representation_vectors_train_only.npy')

# ─── Preprocess (optional but recommended) ───────────────────────────────────
# t-SNE works better on standardized data
dat_scaled = StandardScaler().fit_transform(dat)

# ─── Fit t-SNE ──────────────────────────────────────────────────────────────
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
dat_tsne = tsne.fit_transform(dat_scaled)

# ─── Plot Results ───────────────────────────────────────────────────────────
plt.figure(figsize=(8, 6))
plt.scatter(dat_tsne[:, 0], dat_tsne[:, 1], s=10, alpha=0.7)
plt.title("t-SNE Visualization of Logistic Regression Representation Vectors")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
